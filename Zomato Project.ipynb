# Step 1- Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Step 2- Create the DataFrame

```python
dataframe = pd.read_csv("C:\\Data Analysis\\Zomato data .csv")
print(dataframe)
```

```python
dataframe
```

# Convert the data type of column - Rate

```python
def handleRate(value):
    value=str(value).split('/')
    value=value[0];
    return float(value)
dataframe['rate']=dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe
```

```python
dataframe.info()
```

# Type of resturant

```python
dataframe.head()
```

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel("type of resturant")
```

# Conclusion - majority of the resturant falls in dinnning category

```python
dataframe.head()
```

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes':grouped_data})
plt.plot(result ,c="purple" , marker="o")
plt.xlabel("Type of restaurant", c="blue", size=20)
plt.ylabel("votes" , c="black" , size=20)
```

# Conclusion - dinning resturants has recieved maximum votes

```python
dataframe.head()
```

```python
plt.hist(dataframe['rate'],bins=5)
plt.title("ratings distribution" )
```

# Conclusion the majority resturants received ratings from 3.5 to 4

# Average order spending by couples

```python
couple_data = dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data)
```

# Which mode receives maximum rating

```python
dataframe.head()
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(x = 'online_order', y='rate', data= dataframe)
```

# Conclusion - offline order received lower rating in comparison to online order

```python
pivot_table = dataframe.pivot_table(index='listed_in(type)',columns='online_order',aggfunc='size',fill_value=0)
sns.heatmap(pivot_table,annot=True, cmap="YlGnBu", fmt='d')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")
plt.show()
```

# Conclusion - Dinning resturants primarily accept offline orders.whereas cafes primarily receive online orders. This suggests that clients prefer to orders in person at resturants, but prefer online ordering at cafes.
